```python
# upload the raw dataset 
data = pd.read_csv(inputpath+'/WV6_Data_csv_v20201117.csv',sep=";",
                  lineterminator='\r',
                  error_bad_lines=False)

print(data.shape)
data.head()
```

```python
# my variables
data['country_name'] = data['V2'].astype('category')
df = data[['V45', 'V46', 'V47', 'V48', 'V49', 'V50', 'V51', 'V52', 'V53', 'V54', 'V55', 'V56', 'V95', 'V96', 'V97', 'V98', 'V99', 'V100', 'V101', 'V102', 'V103', 'V104', 'V105', 'V106', 'V107', 'V108', 'V109', 'V110', 'V111', 'V112', 'V113', 'V114', 'V115', 'V116', 'V117', 'V118', 'V119', 'V120', 'V121', 'V122', 'V123', 'V124', 'V127', 'V128', 'V129', 'V130', 'V131', 'V132', 'V133', 'V134', 'V135', 'V136', 'V137', 'V138', 'V139', 'V140', 'country_name']]
```

```python
# drop NaNs
df = df.dropna()
```

```python
# collapse by mean
by_country = df.groupby('country_name').agg({col: 'mean' for col in df.columns if col.startswith('V')})
print(by_country)
#by_country=by_country.iloc[6:12,2:]
```

```python
# take the transpose
t_by_country = by_country.T
t_by_country = pd.DataFrame(t_by_country)
print(t_by_country)
```

```python
# normalize the data
scaler = StandardScaler()
by_country_scaled = scaler.fit_transform(by_country)
by_country_scaled_df = pd.DataFrame(by_country_scaled,columns = by_country.columns)
#by_country_scaled_df.describe()
print(by_country_scaled_df)
```

```python
# normalize the transposed data
scaler = StandardScaler()
t_by_country_scaled = scaler.fit_transform(t_by_country)
t_by_country_scaled_df = pd.DataFrame(t_by_country_scaled,columns = t_by_country.columns)
#t_by_country_scaled_df.describe()
print(t_by_country_scaled_df)
```

```python
# save the data frames 
by_country.to_stata(outputpath+'/by_country.dta')
t_by_country.to_stata(outputpath+'/t_by_country.dta')
by_country_scaled_df.to_stata(outputpath+'/by_country_scaled.dta')
t_by_country_scaled_df.to_stata(outputpath+'/t_by_country_scaled.dta')
```
