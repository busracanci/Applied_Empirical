```python
# Define the URL of the Numbeo page you want to scrape
url = 'https://www.numbeo.com/cost-of-living/rankings_by_country.jsp'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table containing cost of living data
    table = soup.find('table', {'id': 't2'})

    # Create empty lists to store data
    categories = []
    prices_usd = []
    countries = []
    cost_of_living_relative = []

    # Iterate through rows in the table
    for row in table.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        categories.append(columns[1].text.strip())
        prices_usd.append(columns[2].text.strip())
        countries.append(columns[0].text.strip())
        cost_of_living_relative.append(columns[3].text.strip())

    # Create a DataFrame to store the data
    data = {
        'Country': countries,
        'Product Category': categories,
        'Price (USD)': prices_usd,
        'Cost of Living Relative to US': cost_of_living_relative
    }
    df = pd.DataFrame(data)

    # Save the data to a CSV file in the specified folder
    csv_file_path = os.path.join(outputpath, 'cost_of_living.csv')
    df.to_csv(csv_file_path, index=False)

    print(f"Data has been scraped and saved to '{csv_file_path}'.")

    # Download and save the page as .html
    html_url = 'https://www.numbeo.com/cost-of-living/'
    html_response = requests.get(html_url)
    if html_response.status_code == 200:
        html_file_path = os.path.join(outputpath, 'numbeo_cost_of_living.html')
        with open(html_file_path, 'w', encoding='utf-8') as html_file:
            html_file.write(html_response.text)
        print(f"Numbeo cost of living page saved as '{html_file_path}'.")

        # Iterate over all countries
        for country_name in countries:
            # Handle country names with spaces by replacing with hyphens and making the URL
            country_url = f'https://www.numbeo.com/cost-of-living/country_result.jsp?country={country_name.replace(" ", "-")}&displayCurrency=USD'

            country_response = requests.get(country_url)
            if country_response.status_code == 200:
                # Load the country HTML file into a BeautifulSoup object
                country_soup = BeautifulSoup(country_response.content, 'html.parser')

                # Find the relevant table
                country_table = country_soup.find('table', {'id': 't2'})

                # Extract the cost of living relative to Sweden using regular expressions
                sweden_relative_cost = re.search(r'Cost of Living in Sweden: ([0-9.]+)%', country_soup.text)
                if sweden_relative_cost:
                    sweden_relative_cost = sweden_relative_cost.group(1)
                else:
                    sweden_relative_cost = 'N/A'

                # Print the table headings, values, and Sweden's cost of living relative
                output_file_path = os.path.join(outputpath, f'{country_name.replace(" ", "_")}_data.txt')
                with open(output_file_path, 'w', encoding='utf-8') as output_file:
                    output_file.write("Table Headings: ")
                    output_file.write(','.join(categories) + '\n')
                    output_file.write("Table Values: ")
                    output_file.write(','.join(prices_usd) + '\n')
                    output_file.write(f"Cost of Living Relative to Sweden: {sweden_relative_cost}%")

                print(f"Country data for '{country_name}' has been saved to '{output_file_path}'.")

            else:
                print(f"Failed to download the country HTML file for '{country_name}'.")

    else:
        print("Failed to download the HTML page.")

else:
    print("Failed to retrieve data from the website.")
```

```python
# Initialize an empty list to store DataFrames
dataframes = []

# Iterate over files in the directory
for filename in os.listdir(outputpath):
    if filename.endswith('.txt'):
        # Construct the full path to the text file
        file_path = os.path.join(outputpath, filename)

        # Read the text file into a DataFrame
        df = pd.read_csv(file_path, sep=',')
        
        # Add the country name to the DataFrame
        country_name = filename.replace('_data.txt', '').replace('_', ' ')
        df['Country'] = country_name

        dataframes.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dataframes, ignore_index=True)

# Convert the combined DataFrame to a Stata .dta file
stata_file_path = '/path/to/your/folder/all_countries_data.dta'  # Change this to your desired output path

combined_df.to_stata(stata_file_path, write_index=False)

print(f"Data has been combined and saved to '{stata_file_path}' as a Stata dataset.")
```
